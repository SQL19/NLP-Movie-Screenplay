This project is to classify the mood of each scene by using Facebook's bart-large-mnli model. Due to time limitation, no extra fine tunning is conducted here.

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('scenes.csv')
df.head()
```

```python
from transformers import pipeline
classifier = pipeline("zero-shot-classification",
                      model="facebook/bart-large-mnli")
```

```python
# specify the labels of sceen mood
scene_mood = ['Serious', 'Comedic', 'Romantic', 'Suspenseful', 'Cheerful', 'Sad',
'Mysterious', 'Calm', 'Hopeful', 'Angry', 'Fearful', 'Tense', 'Lonely']
```

```python
# to speed computation time, here just generate scene mood for the first 5 scenes
df['scene_mood'] = ''
for i in range(len(df))[:5]:
    scene = df.loc[i, 'text']
    mood = classifier(scene, scene_mood)
    df.loc[i, 'scene_mood'] = mood['labels'][0]
```
